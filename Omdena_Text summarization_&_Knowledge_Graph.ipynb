Text CLeaning

```
import nltk
nltk.download('all')
```

```
text = XLNet_summary


from nltk.tokenize import sent_tokenize
tokens = sent_tokenize(text)
print(tokens)
```

```
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string


stop_words = set(stopwords.words('english'))
stop_words.add('The')


print(stop_words)
```

```
list_of_sentences = []

for sent in tokens:
  word_tokens = word_tokenize(sent)
  words_no_lower = [w.lower() for w in word_tokens]
  #stripped = [w.translate(table) for w in words_no_lower]
  words_no_punct = [word for word in words_no_lower if word.isalpha()]
  words_no_stopw = [w for w in word_tokens if not w in stop_words]
  sentence = ' '.join(word for word in words_no_stopw)
  list_of_sentences.append(sentence)


print(list_of_sentences)
clean_text = ' '.join(s for s in list_of_sentences)
print(clean_text)
```

**Chunking Sentences**

```
from allennlp.predictors.predictor import Predictor
import allennlp_models.structured_prediction
predictor = Predictor.from_path("https://storage.googleapis.com/allennlp-public-models/openie-model.2020.03.26.tar.gz")
x = predictor.predict(
  sentence=clean_text
)

## output
print('AllenNLP Outcome: ')
print()
print(x)
```

```
match_list = []

for v in x['verbs']:
  desc = v['description']
  matches = re.findall(r'\[.*?\]', desc)
  match_list.append(matches) 
  print(matches)
  

print(match_list)
```

```
source_list = []

for i in match_list:
  source_dict = {}
  for j in i:
    pattern1 = re.compile(r'[A-Z+\d]+:|[\w*]+-+[\w*]+:')
    pattern2 = re.compile(r':(.+)\]$')
    result1 = pattern1.findall(j)
    result2 = pattern2.findall(j)
    source_dict.update({result1[0] : result2[0]})
  print(source_dict)
  print(type(source_dict))
  source_list.append(source_dict)

print(source_list)
```
